```python
# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
```

````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        while True:
            try:
                quantity = int(input(f"Enter quantity for {product}: "))
                if quantity <= 0:
                    raise ValueError
                inventory[product] = quantity
                break
            except ValueError:
                print("Error: you must enter a positive value for an order.")
    return inventory


def calculate_total_price(customer_orders):
    total = 0
    for product in customer_orders:
        while True:
            try:
                price = float(input(f"Enter price for {product}: "))
                if price <= 0:
                    raise ValueError
                total += price
                break
            except ValueError:
                print("Error: you must enter a positive value for price.")
    return total


def get_customer_orders(inventory):
    orders = []
    while True:
        try:
            n = int(input("Enter number of orders: "))
            if n <= 0:
                raise ValueError
            break
        except ValueError:
            print("Error: you must enter a positive value for an order.")

    for _ in range(n):
        while True:
            product = input("Enter product name: ")
            if product not in inventory:
                print("Error: you must enter a valid product name.")
            elif inventory[product] <= 0:
                print("Error: Product out of stock.")
            else:
                orders.append(product)
                break
                
    return orders


products = ["apple", "banana", "orange"]
inventory = initialize_inventory(products)
orders = get_customer_orders(inventory)
total_price = calculate_total_price(orders)

print(inventory)
print(orders)
print(total_price)
````
